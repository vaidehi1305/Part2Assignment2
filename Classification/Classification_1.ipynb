```python
def addVantageToFico(df_input):
    ceil_function= lambda x: math.ceil(x)
    df_input['Fico_Score'] = (df_input['Risk_Score']-500)*1.124 + 299
    df_input['Fico_Score']=df_input['Fico_Score'].apply(ceil_function)
    return df_input
```
